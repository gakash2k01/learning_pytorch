```python
import numpy as np
import pandas as pd
import torch
```

```python
a = torch.tensor([5.], requires_grad=True)
b = torch.tensor([6.], requires_grad=True)
```

```python
y = a**3 - b**2
```

```python
y
```

```python
# dy/da = 3*(a**2) = 75
# dy/db = -2*(b) = -12
```

```python
print(a.grad)
```

```python
print(b.grad)
```

```python
y.backward()
```

```python
a.grad
```

```python
print(b.grad)
```

```python
W = torch.randn(10, 1, requires_grad=True)
b = torch.randn(1, requires_grad=True)
```

```python
W
```

```python
x = torch.rand(1,10)
```

```python
x
```

```python
torch.matmul(x, W) + b
```

```python
output = torch.matmul(x,W) + b
```

```python
loss = 1 - output
```

```python
output
```

```python
loss
```

```python
loss.backward()
```

```python
print(W.grad)
```

```python
print(b.grad)
```

```python
learning_rate = 0.001
```

```python
with torch.no_grad():
  W = W - learning_rate * W.grad.data
```
